## Implementação de um Perceptron Simples

```python
x = [1 ,2, 3]  #Entradas dada nos Exercicios
w = [0.1, 0.2, 0.3] #pesos de casa entrada
b = 0 #bias 0 sem importancia
```

```python
x[0]*w[0] # Testando as multiplicaçoes
x[1]*w[1]
x[2]*w[2]
```

```python
fsoma = x[0]*w[0] + x[1]*w[1] + x[2]*w[2] + 0
fsoma
# Foi feito o calculo da função soma dados os dados a cima com bias 0
```

```python
def soma(entradas, pesos, bias):
  soma = 0
  for i in range(len(entradas)):
    soma += entradas[i] * pesos[i]
  soma +=  bias
  return soma
# essa função esta começando no valor 0 e cada entrada que temos esta sendo multiplicado pelo peso e somado pelo bias.
```

```python
u = soma(x, w, b)
u
```

```python
def funcao_degrau(u):
  if u >= 0:
    return 1
  else:
    return
    # Definir a função de ativação: FUNÇÃO DEGRAU
```

```python
u = soma(x, w, b)
y = funcao_degrau(u)

print(f"Para as entras {x} e pesos {w} com bias = {b} a soma foi {u} e o percepetron dado é {y}")
```

##.loc

- No pandas, loc é um método utilizado para acessar linhas e colunas de um DataFrame usando rótulos. Ele é particularmente útil quando você está trabalhando com índices não numéricos ou quando deseja acessar dados por nome de coluna.

- Exemplo Suponha que você tenha um DataFrame df como este:       Nome     Idade
ID                   
1      Alice    23
2      Bob      30
3      Charlie  25


- Acesso a uma Linha: df.loc[3] retornaria a linha inteira onde o índice é 3:
Nome     Charlie
Idade    25
Name: 3, dtype: object



- Acesso a um Elemento Específico: Você pode acessar um elemento específico usando df.loc[3, 'Nome'], que retornaria "Charlie".


- Acesso a Múltiplas Linhas ou Colunas: Também é possível acessar múltiplas linhas ou colunas de uma vez, como df.loc[[1, 3], ['Nome', 'Idade']], que retornaria um DataFrame com as linhas para IDs 1 e 3 e as colunas "Nome" e "Idade".
