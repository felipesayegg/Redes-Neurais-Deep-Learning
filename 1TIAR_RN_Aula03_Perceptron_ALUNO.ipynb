## Implementação de um Perceptron simples

```python
X = [1,2,3] #entradas
W = [0.1, 0.2, 0.3] #pesos
b = 0 #bias
```

```python
X[2] * W[2]
```

```python
X[0] * W[0] + X[1] * W[1] + X[2] * W[2] + b
```

```python
# Definir a função de agregação de informação (soma poderada)
def soma(entradas, pesos, bias):
    soma = 0
    for i in range(len(entradas)):
        soma += entradas[i] * pesos[i]
    soma += bias
    return soma
```

```python
#u = soma(X, W , b)
```

```python
# Definir a função de ativação: FUNÇÃO DEGRAU
def funcao_degrau(u):
  if u>=0:
    return 1
  else:
    return 0
```

```python
u = soma(X, W , b)
y = funcao_degrau(u)

print( f'Para as entradas X = {X} e pesos W = {W}, a soma foi {u} e o perceptron foi {y}')
```

## Casos de aplicação do perceptron

Vamos analisar a base `escolha_faculdade.xlsx` com exemplos de faculdades identificadas pelo ID e com um conjunto de variáveis que as descrevem. As variáveis são:

- reconhecida_MEC: a faculdade é reconhecida pelo MEC? 1=sim ; 0=não
- facilidade_pagamento: a faculdade apresenta facilidades para pagamento? 1=sim ; 0=não
- flexibilidade_ensino: o quão flexível é o método de ensino? 1=somente presencial ; 2=somente EAD ; 3=misto
- entrada_mercado: o que a faculdade oferece para facilitar a entrada do aluno no mercado de trabalho? 0=sem suporte ; 1=feiras ; 2=feiras e estágios ; 3=feiras, estágios e convênios com empresas
- score_infraestrutura: qual nota de 0 a 100 a escola possui para sua infraestrutura?
- empregabilidade: qual a porcentagem de alunos egressos empregados?
- acesso_facil: a localização possui fácil acesso, como metrô, ônibus, estacionamento? 1=sim ; 0=não

```python
# Pesos já fornecidos
pesos = [ 0.10, 0.05, 0.05, 0.2, 0.25, 0.3, 0.05 ]
```

```python
import pandas as pd # pre processamento de conjuntos
import math # funcoes matematicas
```

```python
df = pd.read_excel("/content/escolha_faculdade.xlsx", index_col="ID")
```

```python
# aqui estamos invertendo a matriz de dados

escolas = df.T
```

```python
# Definir a função de agregação de informação (soma poderada)
def soma(entradas, pesos, bias):
    soma = 0
    for i in range(len(entradas)):
        soma += entradas[i] * pesos[i]
    soma += bias
    return soma
```

```python
u1 =soma(escolas[1],pesos, 0)
```

```python
#Definir a funcao de ativacao (funcao degrau) para o problema de classificacao
# Função degrau com threshold =35

def funcao_degrau(u):
  if u  >=35:
    return 1
  else:
    return 0
```

```python
u8 =soma(escolas[8],pesos, 0)
u8
```

```python
funcao_degrau(u8)
```

```python
# Definindo a funcao de ativacao para a regressao : Funcao Identidae elevada ao quadrado

def funcao_identidade(u):
  return u
```

```python
score8 = funcao_identidade(u8)
score8
```

```python
score9 = funcao_identidade(u9)
score9
```

```python
resp_class = [] # resultado da classificacao (lista vazia)
resp_reg = []   # resultado da regressao (lista vazia)
for i in range(1,11): # para todo numeor i de 1 até 10 faça:
  e = escolas[i]  #obter a escola de ID i

  u = soma(e, pesos, 0) # valor da soma ponderada

  classificacao = funcao_degrau(u) # valor da classificacao
  score = funcao_identidade(u) # valor da regressao

  resp_class.append(classificacao) # adicionar a classificacao na lista
  resp_reg.append(score) # adicionar a regressao na lista

```

```python
resp_class
```

```python
resp_reg
```

```python
df["APROVADO"] = resp_class
df["NOTA_ESCOLA"] = resp_reg
```
